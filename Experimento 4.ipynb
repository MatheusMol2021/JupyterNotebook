```python
indvíduo = []
classe = []
classe_numero = []
anomalia = []
listaBPM = []

for diretorio, subpastas, arquivos in tqdm(os.walk(dir_data)):
    for arquivo in arquivos:
        
        # Lista o Arquivo
        #print(arquivo)
        
        # Nome do indivíduo
        b = arquivo.replace(".csv","")                           # Define o nome do indivíduo, de acordo com o arquivo
        indvíduo.append(b)
        c = dfclas.loc[dfclas['FileName'] == b]                  # Procura a linha que tem os dados do indivído
        d = str(c['Rhythm'])                                     # Procura a coluna com os dados do indivíduo
        e = d.split()                                            # Separa os dados em strings
        #print(e[1])                                              # imprime somente o indice que tem a classificação
        classe.append(e[1])
        
        # Classificação binaria anomalia = 1, normal = 0
        if e[1] == 'SN':
            #print('1')
            anomalia.append(1)
        else:
            #print('0')
            anomalia.append(0)
        
        # Classificação anoalia para inteiros
        if e[1] == 'SB': 
            # Sinus Bradycardia
            classe_numero.append(1)
        elif e[1] == 'SR':
            # Sinus Rhythm
            classe_numero.append(2)
        elif e[1] == 'AFIB':
            # Atrial Fibrillation
            classe_numero.append(3)
        elif e[1] == 'ST':
            # Sinus Tachycardia
            classe_numero.append(4)
        elif e[1] == 'AF':
            # Atrial Flutter
            classe_numero.append(5)
        elif e[1] == 'SI':
            # Sinus Irregularity
            classe_numero.append(6)
        elif e[1] == 'SVT':
            # Supraventricular Tachycardia
            classe_numero.append(7)
        elif e[1] == 'AT':
            # Atrial Tachycardia
            classe_numero.append(8)
        elif e[1] == 'AVNRT':
            # Atrioventricular  Node Reentrant Tachycardia
            classe_numero.append(9)
        elif e[1] == 'AVRT':
            # Atrioventricular Reentrant Tachycardia
            classe_numero.append(10)
        elif e[1] == 'SAAWR':
            # Sinus Atrium to Atrial Wandering Rhythm
            classe_numero.append(11)      
        else:
            classe_numero.append(20)
        
        # Converte BPM
        a = os.path.join(diretorio, arquivo)                      # Variável com o nome do Diretório do arquivo
        df = pd.read_csv(a)                                       # Abre o arquivo do Diretório
        y = df["II"].values                                       # Valores para y correspondem a Temperatura aparente
        out = ecg.ecg(signal=y, sampling_rate=500., show=False)   # Insere os dados na biblioteca SSPY
        z = out['heart_rate']                                     # Extrai os dados em BPM
        listaBPM.append(z)                                        # Adiciona od dados BPM em uma lista
       
"""        s1 = pd.Series([b, classe, anomalia, z])

        df = pd.DataFrame([list(s1)],  columns =  ["Nome", "Classe", "Anomalia", "BPM"])
        df
"""

listaBPM
```

```python
X = listaBPM
X
```

```python
y = classe_numero
y
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train = np.asarray(X_train)
y_train = np.asarray(y_train)
X_test = np.asarray(X_test)
y_test = np.asarray(y_test)
```

```python

import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing import sequence
# fix random seed for reproducibility
tf.random.set_seed(7)
```

```python

# truncate and pad input sequences
max_review_length = 16
X_train = sequence.pad_sequences(X_train, maxlen=max_review_length)
X_test = sequence.pad_sequences(X_test, maxlen=max_review_length)
```

```python
X_train[15]
```

```python
# create the model
top_words = 200
embedding_vecor_length = 5
model = Sequential()
model.add(Embedding(top_words, embedding_vecor_length, input_length=max_review_length))
model.add(LSTM(10))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=3, batch_size=64)
```

```python

# Final evaluation of the model
scores = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```
